```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
from matplotlib.axes import Axes
from pandas import DataFrame
import seaborn as sns
```

'FILE.KEGG.HSJD.csv' DataFrame

```python
data_frame: DataFrame
data_frame = pd.read_csv(filepath_or_buffer='FILE.KEGG.HSJD.csv', delimiter=',')
```

Ulcerative colitis DataFrame

```python
cu: DataFrame
cu = pd.concat(objs=[data_frame[data_frame['GROUP'].isin([
    'fecal.Control.control_vs_fecal.CU.3.meses',
    'fecal.Control.control_vs_fecal.CU.6.meses',
    'fecal.Control.control_vs_fecal.CU.debut',
    'fecal.CU.6.meses_vs_fecal.CU.debut'
])]])
cu = cu.sort_values(by='EA_VALUE', ascending=False)
```

Crohn's disease DataFrame

```python
ec: DataFrame
ec = pd.concat(objs=[data_frame[data_frame['GROUP'].isin([
    'fecal.Control.control_vs_fecal.EC.6.meses',
    'fecal.EC.3.meses_vs_fecal.Control.control',
    'fecal.EC.3.meses_vs_fecal.EC.6.meses',
    'fecal.EC.3.meses_vs_fecal.EC.debut',
    'fecal.EC.debut_vs_fecal.EC.6.meses'
])]])
ec = ec.sort_values(by='EA_VALUE', ascending=False)
```

Ulcerative colitis plot

```python
fig: Figure
ax: Axes
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

for index, (ax, (group, data)) in enumerate(zip(ax.flatten(), cu.groupby('GROUP'))):
    sns.swarmplot(data=data, x='KEGG_up_DOWN', y='ONTOLOGY', hue='EA_VALUE', ax=ax, size=10)
    ax.set_title(label=group)
    ax.legend(bbox_to_anchor=(1, 1), loc='upper left', labels=data['EA_VALUE'].unique(), title='EA_VALUE')
    
plt.subplots_adjust(wspace=1, hspace=0.3)
```

Crohn's disease plot

```python
fig: Figure
ax: Axes
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

for index, (ax, (group, data)) in enumerate(zip(ax.flatten(), ec.groupby('GROUP'))):
    sns.swarmplot(data=data, x='KEGG_up_DOWN', y='ONTOLOGY', hue='EA_VALUE', ax=ax, size=10)
    ax.set_title(label=group)
    ax.legend(bbox_to_anchor=(1, 1), loc='upper left', labels=data['EA_VALUE'].unique(), title='EA_VALUE')
    
plt.subplots_adjust(wspace=1, hspace=0.3)
```
